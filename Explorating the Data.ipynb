### Survivors/Victims analysis

The previous section studied only the survivors of the Titanic, which showed up some information about their demographics. We will now have a look at the survivors/victims demographic, which will show other useful information and give us a feeling of potential hidden patterns.

````python
# Split the dataset as survivors/victims
survivors = df[df.Survived == 1]
victims = df[df.Survived == 0]
````

````python
survivors.head()
````

````python
victims.head()
````

As we did previousely, we will start by comparing survivors and victims based on their ages. Remember that ~20% of the `age` attribute information are missing.

````python
age_ranges = np.arange(0, 90, 10)
survivors_per_age_range = [df[(df['Survived'] == 1) & (df['Age'] >= a) & (df['Age'] < a + 10)] for a in age_ranges]
victims_per_age_range = [df[(df['Survived'] == 0) & (df['Age'] >= a) & (df['Age'] < a + 10)] for a in age_ranges]

len_survivors = [len(survivor) for survivor in survivors_per_age_range]
len_victims = [len(survivor) for survivor in victims_per_age_range]
````

````python
p1 = plt.bar(age_ranges, len_survivors, width = 9, color = 'b')
p2 = plt.bar(age_ranges, len_victims, width = 9, color = 'r', bottom = len_survivors)

plt.legend((p2, p1), ('Victims', 'Survivors'))
plt.xlabel('Age of passengers')
plt.ylabel('Count')
plt.title('Count of survivors and victims based on ages')
plt.show()
````

Except for the 0-10 years old range, the total of victims for each age range is higher than the total of survivals, with a peak for the 20-30 years old range. This graph shows the influence of age in the survival.

Plotting the survival bar graph depending on the sex of passenger.

````python
survivors_per_sex = [len(survivors[survivors.Sex == sex]) for sex in ['female', 'male']]
victims_per_sex = [len(victims[victims.Sex == sex]) for sex in ['female', 'male']]

p1 = plt.bar([0, 1], survivors_per_sex, color = 'b')
p2 = plt.bar([0, 1], victims_per_sex, color = 'r', bottom = survivors_per_sex)

plt.xticks([0.4, 1.4], ('Female', 'Male'))
plt.legend((p2, p1), ('Victims', 'Survivors'))
plt.ylabel('Count')
plt.title('Count of survivors/victims based on Sex attribute')
plt.show()
````

This gives the information that women are more likely to survive than men in this kind of tragedy.

Let's now do the same analysis with the passenger classes.

````python
survivors_per_class = [len(survivors[survivors.Pclass == pclass]) for pclass in [1, 2, 3]]
victims_per_class = [len(victims[victims.Pclass == pclass]) for pclass in [1, 2, 3]]

p1 = plt.bar([0, 1, 2], survivors_per_class, color = 'b')
p2 = plt.bar([0, 1, 2], victims_per_class, color = 'r', bottom = survivors_per_class)

plt.xticks([0.4, 1.4, 2.4], ('1st Class', '2nd Class', '3rd Class'))
plt.legend((p2, p1), ('Victims', 'Survivors'))
plt.ylabel('Count')
plt.title('Count of survivors/victims based on passenger class attribute')
plt.show()
````

The display of the passengers survival based on the travel fare is better visualized through scatterplotting.

````python
survivors_ticket_fare = survivors.Fare
victims_ticket_fare = victims.Fare

plt.scatter(survivors_ticket_fare, np.ones(len(survivors_ticket_fare)), marker = 'x', color = 'green', alpha = 0.5)
plt.scatter(victims_ticket_fare, np.zeros(len(victims_ticket_fare)), marker = 'o', color = 'red', alpha = 0.5)
plt.xlabel('Ticket Fare')
plt.ylabel('0- Victim | 1- Survivor')
plt.title('Survivors/Victims based on ticket fare')
plt.show()
````

Difficult to say from the ticket fare who survived and who died. This was more clear from the passenger class and sex.

We will now scatter some plots showing other attributes.

````python
plt.subplot(211)
plt.scatter(survivors_ticket_fare, survivors.Pclass, marker = 'x', color = 'green', alpha = 0.5)
plt.xlabel('Ticket fare')
plt.ylabel('Survivor Class')

plt.subplot(212)
plt.scatter(victims_ticket_fare, victims.Pclass, marker = 'o', color = 'red', alpha = 0.5)
plt.xlabel('Ticket fare')
plt.ylabel('Victim Class')

plt.show()
````

Still not relevant to visually predict passenger survival depending on the ticket fare and passenger class. Even if we can see that survivors of 3rd class passengers is lower that victims, this is best visualized using the previous bar chart.

## Features Correlations

Let's now have a look at features correlation to have an idea on variables (in)dependance.

````python
df.corr()
````

Looking at the `Survived` feature, which represents the survival, we can say that it is more correlated to the passenger class `Pclass` feature and the ticket fare `Fare` feature.

We can see that the passenger sex is not represented in the correlation matrix, because `Sex` feature is a string and not a numerical value. So, instead of looking at the whole dataframe as extracted from the dataset, let's first create a new dataframe with numerical values for `Sex`, and remove some unnecessary features (like `PassengerId`).

````python
df2 = df

# Drop unnecessary features for correlation
df2 = df2.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 1)

# Create dummy variables for string features
dummy_sex = pd.get_dummies(df2.Sex, prefix = 'Sex')
dummy_embarked = pd.get_dummies(df2.Embarked, prefix = 'Embarked')

# Drop original Sex and Embarked features & join dummy variables to new dataframe 
df2 = df2.drop(['Sex', 'Embarked'], axis = 1)
df2 = df2.join(dummy_sex)
df2 = df2.join(dummy_embarked)

df2.head()
````

````python
df2.corr()
````

Now we have a better idea on what features the `Survived` attribute directly depends on.

````python
df2.corr().Survived
````

The passenger survival is more demendant on the passenger class `Pclass`, the ticket fare `Fare` and the passenger sex. The correlation shows what variables directly affect the survival.
