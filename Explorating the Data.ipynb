## Some analysis to understand the data and the context

Here we will ask a serie of questions to better understand the data and get familiar with them (and have some fun...)

**Q: How was the survival rate among passengers of age 60 and more?**

````python
df[df['Age'] >= 60].Survived.mean()
````

The survival rate dropped to ~27% for people of 60 years old and more. Here is a histogram of survival among ages:

````python
survivors = df[df.Survived == 1]
survivors.head()
````

````python
age_ranges = np.arange(0, 90, 10)
survivors_per_age_range = [df[(df['Survived'] == 1) & (df['Age'] >= a) & (df['Age'] < a + 10)] for a in age_ranges]

survivors_per_age_range
#age_ranges
````

Let's see the distribution of survivors among age ranges of 10 years:

````python
len_survivors = [len(survivor) for survivor in survivors_per_age_range]
plt.bar(age_ranges, len_survivors, width = 9)
plt.show()

print len_survivors
````

More information about the last bar chart, showing Sex of survivors:

````python
len_survivors_women = [len(survivor[survivor.Sex == 'female']) for survivor in survivors_per_age_range]
len_survivors_men = [len(survivor[survivor.Sex == 'male']) for survivor in survivors_per_age_range]
p1 = plt.bar(age_ranges, len_survivors_women, width = 9, color = 'r')
p2 = plt.bar(age_ranges, len_survivors_men, width = 9, color = 'b', bottom = len_survivors_women)
plt.legend((p1, p2), ('Women', 'Men'))
plt.show()

print len_survivors_women
print len_survivors_men
````

Same as before, but with embarkment information:

````python
len_survivors_s = [len(survivor[survivor.Embarked == 'S']) for survivor in survivors_per_age_range]
len_survivors_c = [len(survivor[survivor.Embarked == 'C']) for survivor in survivors_per_age_range]
len_survivors_q = [len(survivor[survivor.Embarked == 'Q']) for survivor in survivors_per_age_range]

p1 = plt.bar(age_ranges, len_survivors_s, width = 9, color = 'b')
p2 = plt.bar(age_ranges, len_survivors_c, width = 9, color = 'r', bottom = len_survivors_s)
p3 = plt.bar(age_ranges, len_survivors_q, width = 9, color = 'g', bottom = [sum(x) for x in zip(len_survivors_s, len_survivors_c)])

plt.legend((p1, p2, p3), ('Southampton', 'Cherbourg', 'Queenstown'))
plt.show()
````

````python
len_survivors_class_1 = [len(survivor[survivor.Pclass == 1]) for survivor in survivors_per_age_range]
len_survivors_class_2 = [len(survivor[survivor.Pclass == 2]) for survivor in survivors_per_age_range]
len_survivors_class_3 = [len(survivor[survivor.Pclass == 3]) for survivor in survivors_per_age_range]

p1 = plt.bar(age_ranges, len_survivors_class_1, width = 9, color = 'b')
p2 = plt.bar(age_ranges, len_survivors_class_2, width = 9, color = 'r', bottom = len_survivors_class_1)
p3 = plt.bar(age_ranges, len_survivors_class_3, width = 9, color = 'g', bottom = [sum(x) for x in zip(len_survivors_class_1, len_survivors_class_2)])

plt.legend((p1, p2, p3), ('1st Class', '2nd Class', '3rd Class'))
plt.show()
````

That doesn't give so much info. Let's now include victims in our analysis.

````python
# Split the dataset as survivors/victims
survivors = df[df.Survived == 1]
victims = df[df.Survived == 0]
````

````python
survivors.head()
````

````python
victims.head()
````

Plotting the survival bar graph depending on the sex of passenger.

````python
survivors_per_sex = [len(survivors[survivors.Sex == sex]) for sex in ['female', 'male']]
victims_per_sex = [len(victims[victims.Sex == sex]) for sex in ['female', 'male']]

p1 = plt.bar([0, 1], survivors_per_sex, color = 'b')
p2 = plt.bar([0, 1], victims_per_sex, color = 'r', bottom = survivors_per_sex)

plt.xticks([0.4, 1.4], ('Female', 'Male'))
plt.legend((p2, p1), ('Victims', 'Survivors'))
plt.show()
````

This gives the information that women are more likely to survive than men in this kind of tragedy.

Let's now do the same analysis with the passenger classes.

````python
survivors_per_class = [len(survivors[survivors.Pclass == pclass]) for pclass in [1, 2, 3]]
victims_per_class = [len(victims[victims.Pclass == pclass]) for pclass in [1, 2, 3]]

p1 = plt.bar([0, 1, 2], survivors_per_class, color = 'b')
p2 = plt.bar([0, 1, 2], victims_per_class, color = 'r', bottom = survivors_per_class)

plt.xticks([0.4, 1.4, 2.4], ('1st Class', '2nd Class', '3rd Class'))
plt.legend((p2, p1), ('Victims', 'Survivors'))
plt.show()
````

The display of the passengers survival based on the travel fare is better visualized through scatterplotting.

````python
survivors_ticket_fare = survivors.Fare
victims_ticket_fare = victims.Fare

plt.scatter(survivors_ticket_fare, np.ones(len(survivors_ticket_fare)), marker = 'x', color = 'green', alpha = 0.5)
plt.scatter(victims_ticket_fare, np.zeros(len(victims_ticket_fare)), marker = 'o', color = 'red', alpha = 0.5)
plt.show()
````

Difficult to say from the ticket fare who survived and who died. This was more clear from the passenger class and sex.

We will now scatter some plots showing other attributes.

````python
plt.subplot(211)
plt.scatter(survivors_ticket_fare, survivors.Pclass, marker = 'x', color = 'green', alpha = 0.5)
plt.title('Survivors scatter plot given the ticket fare and passenger class')

plt.subplot(212)
plt.scatter(victims_ticket_fare, victims.Pclass, marker = 'o', color = 'red', alpha = 0.5)
plt.title('Victims scatter plot given the ticket fare and passenger class')

plt.show()
````

Still not relevant to visually predict passenger survival depending on the tocket fare and passenger class. Even if we can see that survivors of 3rd class passengers is lower that victims, this is best visualized using the previous bar chart.
